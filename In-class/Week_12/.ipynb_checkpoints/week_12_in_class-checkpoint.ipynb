```python
import csv
import matplotlib.pyplot as plt
import re
# shows graphs inline
%matplotlib inline

## Open the file
dog_file = open('dog_tags.csv', newline='')
## Read the file using the csv library DictReader method
reader = csv.DictReader(dog_file)

## print to make sure it works
#for row in reader:
#    print(row['DogName'])

## Use the raw interpretation of regex to match for 
## any and all dog names
#exp = re.compile(r'.*')

## Using the match() function to find matches
#for row in reader:
#    m = exp.match(row['DogName'])
#    print(m)

## Expanding on the example to match all names starting
## with S and print only the matches
#exp = re.compile(r'S.*')

## Additional funcationality added with ignoring case
# exp = re.compile(r'BI.*', re.IGNORECASE)
# for row in reader:
#     m = exp.match(row['DogName'])
#     if m:
#         print(m)
        
## Adding some more complexity to accomodate names
## with a character class (/s)
# exp = re.compile(r'.*\s.*', re.IGNORECASE)
# for row in reader:
#     m = exp.match(row['DogName'])
#     if m:
#         print(m)

## define a new charachter class (using brackets)
## to only pull names starting with X or G
# exp = re.compile(r'[XG].*', re.IGNORECASE)
# for row in reader:
#     m = exp.match(row['DogName'])
#     # checking we don't have a None object
#     if m:
#         # call group() to get full output not objects
#         print(m.group())
        
"""
Worked on some regex practice with same csv using
some peer challenges

https://docs.google.com/spreadsheets/d/1ccfisdttdfuBGl1b6B_39s44J_VWelkeCt-xJDbUDIQ/edit#gid=364569146
"""

## Challenge:
## Dog name starts and ends with the same letter

new_exp = re.compile(r'([a-z]).*\1$', re.IGNORECASE)
for row in reader:
    challenge = new_exp.match(row['DogName'])
    # checking we don't have a None object
    if challenge:
        # call group() to get full output not objects
        print(challenge.group())

"""
group notation with paranthesis saves the regex as a
variable
then you have .* for anything in between
and \1 references the first capturing group 
i.e. [a-zA-Z]. This is called a "back reference"
dollar sign matches the end of a line (inverse of ^)

([a-zA-Z]).*\1$
"""
```

```python
"""
Matplotlib!
Mapping out personal energy levels (y)
versus the time of day (x)

Imagine spreading the time in increments of 4
i.e. 0-4, 4-8, 8-12, 12-16, 16-20, 20-24

Library we're calling is MatPlotLib
which takes in axes objects (x,y)

Underlying the library is a numpy array for each point
on the line we're graphing.
It's using a datadrame with an index object and each
column is a series object.

Dataframe.plot() combines with matplotlib to plot the 
dataframe.

Usage:
https://matplotlib.org/tutorials/introductory/usage.html#backends
"""
## A quick test to check to see how this works
## by passing list objects to plot. First x then y!
#plt.plot([10,15,20,25], [10,100,1000,10000])
## still using the same memory spot so it plots on the same graph
#plt.plot([4,8,12,16], [20,45, 250, 8000])

times = [0, 4, 8, 12, 16, 20, 24]
energy_level = [0, 2, 6, 10, 7, 5, 1]
baby_energy = [10, 10, 10, 10, 10, 10, 10]
plt.plot(times, energy_level, label='Me')
plt.plot(times, baby_energy, label='Baby')
plt.ylabel("How energized am I?")
plt.xlabel("What time is it?")
plt.title("Energry Levels Throughout The Day")
plt.legend()
```
