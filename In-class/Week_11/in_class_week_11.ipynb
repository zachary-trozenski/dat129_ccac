```python
# import statements
#from zipfile import ZipFile # unzipped in terminal
import os
#import shutil # not used tonight
import re

# compiling this so that it can be referenced faster
# find me all matches where the match begins with a dot
# contains any of the [a-zA-Z0-9_] once or more at
# the end of the line
regex = re.compile('(\.\w+)$')

def interrogate_files(directory, file_list):
    """
    Grab a directory name and its contents
    Count files with .jpg suffix
    os.sep identifies operating system path seperators
    Then print the size of the file size in bytes
    Using re.search() find the matches in the regex variable
    for each file name and if they match print the extension
    """
    for files in file_list:
        file_name = directory + str(os.sep) + files
        print('file_name: ', file_name)
        if os.path.isfile(file_name):
            print('file size: ', end='')
            print(os.path.getsize(file_name), 'B')
            match = re.search(regex, file_name)
            if match:
                print('extension: ', match.group(0))
        
```

```python
def traverse_tree(root_to_walk):
    """
    this will take a stroll through the current
    directory and print the location, directory, and
    files
    """ 
    for loc, dir, files in os.walk(root_to_walk):
        print(loc, '  ', dir, '  ', files, ' ')
        interrogate_files(loc, files)
    
```

```python
traverse_tree('.')
```
